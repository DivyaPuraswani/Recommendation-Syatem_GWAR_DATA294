Changing datatype of column number_of_reviews from object to numerical

```python
def mapnumber_of_reviews(v): 
    if pd.isnull(v):
        return 0
    try:
        vv = v.replace(",","")
        return int(vv)
    except ValueError:
        return 0
```

```python
if amazon['number_of_reviews'].dtype != 'int64':  
    amazon['number_of_reviews'] = amazon['number_of_reviews'].map(mapnumber_of_reviews)
```

```python
def recommend_me(item):
    tokens = item.split(" ")
    tokens = [t.lower() for t in tokens]
    df = amazon[amazon.product_name.map(lambda name: all(n in name.lower() for n in tokens))]
    
    if len(df[df['average_review_rating'] >= 3.0]) == 0:
        return df.sort_values(by='average_review_rating').iloc[0]
    else:
        return df[df['average_review_rating'] >= 3.0].sort_values(by='number_of_reviews').iloc[0]
```

```python
rating_crosstab1 = amazon_data1.pivot_table(values='average_review_rating', index='unique_id', columns='product_name', fill_value=0)
rating_crosstab1.head()
```

```python
rating_crosstab = amazon_data.pivot_table(values='average_review_rating', index='unique_id', columns='product_name', fill_value=0)
rating_crosstab.head()
```

```python
rating_crosstab1.shape
```

```python
X = rating_crosstab1.T
X.shape
```

```python
SVD = TruncatedSVD(n_components=12, random_state=17)

resultant_matrix = SVD.fit_transform(X)

resultant_matrix.shape
```

```python
corr_mat = np.corrcoef(resultant_matrix)
corr_mat.shape
```
